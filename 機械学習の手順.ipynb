```python
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.manifold import LocallyLinearEmbedding
from pandas.plotting import scatter_matrix
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = load_wine()
```

```python
X, y = data.data, data.target
```

```python
X_pd = pd.DataFrame(data.data, columns=data.feature_names)
y_pd = pd.DataFrame(data.target, columns=['Species'])
```

```python
df = pd.concat([X_pd, y_pd], axis=1)
```

```python
_ = scatter_matrix(df, figsize=(15, 15))
```

```python
#trans_model = LocallyLinearEmbedding(n_components=5, n_neighbors=11)
trans_model = PCA(n_components=2)
```

```python
trans_model.fit_transform(X)
```

```python
model_rfc = RandomForestClassifier()
```

```python
cv = KFold(5, shuffle=True)
```

```python
params_grid = {'max_depth': [5, 10, 15], 'n_estimators':[10, 20, 30]}
```

```python
grid_search = GridSearchCV(model_rfc, params_grid, cv=cv, scoring='accuracy')
```

```python
grid_search.fit(X, y)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
```
